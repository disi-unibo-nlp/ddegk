```python
import json
import networkx as nx
import numpy as np
```

```python
folder = 'st09'
```

```python
fn = 'stratified_samples/' + folder + '_stratified_sample.json'
```

```python
with open(fn) as ff:
    graphs = [nx.node_link_graph(g) for g in json.load(ff)]
```

```python
adj_matrix_path = 'graph_2D_CNN/datasets/data_as_adj/'+folder+'/'
```

```python
for i in range(len(graphs)):
    g = graphs[i]
    m = nx.convert_matrix.to_numpy_matrix(g).astype(int)
    with open(adj_matrix_path+str(i)+'.txt', 'w') as f:
        for line in m:
            np.savetxt(f, line, fmt='%i')
```

```python
# classes as event types
with open(fn) as file:
    graphs = [nx.node_link_graph(g) for g in json.load(file)]
    roots = [g.graph['root'] for g in graphs]
    nodes_list = [list(g.nodes.data('type')) for g in graphs]
    names = [n[1] for i in range(len(graphs)) for n in nodes_list[i] if n[0]== roots[i]] 
```

```python
classes = list(set(names))
classes = [classes.index(n) for n in names]
```

```python
path = 'graph_2D_CNN/datasets/classes/'+folder+'/'
with open(path+folder+'_classes.txt', 'w') as f:
    for c in classes:
        f.write(str(c))
        f.write('\n')
```
